```python
# Dependencies
import requests
import pandas as pd
import numpy as np
from config import api_key
from config import youtube_key

# read in csv file for title lis
quick_csv = "../data_files/quick_merge.csv"
```

```python
# read in csv file for title list
quick_df = pd.read_csv(quick_csv)
quick_df.drop(columns='Unnamed: 0',inplace=True)
quick_df.drop(columns='Ranking per year',inplace=True)
quick_df['trailer ids']=''
quick_df['max viewcount']=''
quick_df.head()
```

```python
# create a title list to search and a list to store trailers
titles = quick_df['Title'].tolist()
tmdb_video_list=[]
```

```python
# search url for tmdb to search movie id by title
search_url =f'https://api.themoviedb.org/3/search/movie?api_key={api_key}&language=en-US&page=1&include_adult=True&query='
i=0
for mov in titles:
    mov_search = requests.get(search_url + mov).json()
    try:
        tmdb_movie_id = mov_search['results'][0]['id']
    except:
        print("movie not found")
        print(mov)
    # get video info with movie id
    video_url = f'https://api.themoviedb.org/3/movie/{tmdb_movie_id}/videos?api_key={api_key}&language=en-US'
    vid_search = requests.get(video_url).json()
    try:
        temp_vid_list=[]
        for vid in vid_search['results']:
            if (vid['type']=='Trailer') & (vid['site']=='YouTube'):
                temp_list.append(vid['key'])
        tmdb_video_list.append(temp_list)
        # Store a list of youtube trailer ids into a column
        quick_df.at[i,'trailer ids']=temp_vid_list
    except:
        print("video not found")
        print(mov)
    i+=1
tmdb_video_list
```

```python
# A forloop for grabbing youtube view counts

for index, row in quick_df.iterrows():
    id_list=row['trailer ids']
    if len(id_list)!=0:
        temp_views_list=[]
        for vid in id_list:
            response = requests.get(f'https://www.googleapis.com/youtube/v3/videos?part=statistics&id={vid}&key={youtube_key}').json()
            try:
                temp_views_list.append(int(response['items'][0]['statistics']['viewCount']))
            except:
                print('youtube call failed')
                print(row['Title'])
        if len(temp_views_list)!=0:
            quick_df.at[index,"max viewcount"]=max(temp_views_list)
        
```
